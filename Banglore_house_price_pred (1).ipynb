```python
def convertRange(x):
  temp = x.split('-')
  if len(temp)==2:
    return (float(temp[0])+float(temp[1]))/2
  try:
    return float(x)
  except:
    return None
```

```python
data['total_sqft'] = data['total_sqft'].apply(convertRange)
```

# **Price per feet**

```python
data['Price per feet'] = data['price']*100000/data['total_sqft']
```

```python
data.head()
```

# **Reducing the location count**

```python
data['location'] = data['location'].apply(lambda x:x.strip())
location_count = data['location'].value_counts()
```

```python
location_count_less_than_10 = location_count[location_count<=10]
location_count_less_than_10
```

```python
data['location'] = data['location'].apply(lambda x: 'other' if x in location_count_less_than_10 else x)
```

# Removing **Outliers**

```python
(data['total_sqft']/data['bhk']).describe()
```

# removing who have less than **300sqft**

```python
data = data[((data['total_sqft']/data['bhk'])>=300)]
data.describe()
```

```python
data['price_per_sqft']=data['Price per feet']
```

```python
def remove_outliers(data):
  df_output = pd.DataFrame()
  for key, subdf in data.groupby('location'):
      m = np.mean(subdf.price_per_sqft)
      st = np.std(subdf.price_per_sqft)

      gen_df = subdf[(subdf.price_per_sqft>(m-st)) & (subdf.price_per_sqft <= (m+st))]
      df_output = pd.concat([df_output, gen_df], ignore_index=True)
  return df_output
data = remove_outliers(data)
data.describe()
```

```python
def bhk_outliers_remover(data):
  exclude_indices = np.array([])
  for location, location_df in data.groupby('location'):
      bhk_stats = {}
      for bhk, bhk_df in location_df.groupby('bhk'):
          bhk_stats[bhk] = {
              'mean':np.mean(bhk_df.price_per_sqft),
              'std':np.std(bhk_df.price_per_sqft),
              'count':bhk_df.shape[0]
          }
      for bhk, bhk_df in location_df.groupby('bhk'):
        stats = bhk_stats.get(bhk-1)
        if stats and stats['count']>5:
            exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft<(stats['mean'])].index.values)
  return data.drop(exclude_indices, axis='index')
```

```python
data = bhk_outliers_remover(data)
```

```python
X = data.drop(columns=['price'])
y = data['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
X_test, X_train, y_test, y_train = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
```

# Applying **LinearRegression**

```python
column_trans = make_column_transformer((OneHotEncoder(handle_unknown='ignore', sparse_output=False), ['location']),remainder='passthrough')
```

```python
scaler = StandardScaler()
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans, scaler, lr)
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred_lr = pipe.predict(X_test)
```

```python
r2_score(y_test, y_pred_lr)
```

# Appying **Lasso**

```python
lasso = Lasso()
```

```python
pipe = make_pipeline(column_trans, scaler, lasso)
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred_lasso = pipe.predict(X_test)
```

```python
r2_score(y_test, y_pred_lasso)
```

# **Applying Ridge**

```python
ridge = Ridge()
```

```python
pipe = make_pipeline(column_trans, scaler, ridge)
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred_ridge = pipe.predict(X_test)
```

```python
r2_score(y_test, y_pred_ridge)
```

```python
print("No Regularization:", r2_score(y_test, y_pred_lr))
print('Lasso:',r2_score(y_test, y_pred_lasso))
print('Ridge:', r2_score(y_test, y_pred_ridge))
```
